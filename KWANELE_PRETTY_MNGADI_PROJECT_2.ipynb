```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("/content/Dataset for Data Analytics 2.xlsx")

print(df.head())

print(df.shape)

print(df.info())
```

PHASE 1 : DATA AUDIT

```python
print("Rows and columns:", df.shape)

print("\nMissing values:")
print(df.isnull().sum())

print("\nDuplicate records:")
print(df.duplicated().sum())

print("\nData types:")
print(df.dtypes)
```

PHASE 2: DESCRIPTIVE STATISTICS

```python
Existing_numeric_cols = ["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]

print(df[Existing_numeric_cols].describe())
```

```python
print("Mean Total Price:", df["TotalPrice"].mean())
print("Median Total Price:", df["TotalPrice"].median())
print("Highest Total Price:", df["TotalPrice"].max())
print("Lowest Total Price:", df["TotalPrice"].min())
```

PHASE 3: DISTRIBUTION ANALYSIS

```python
plt.hist(df["TotalPrice"], bins=20,edgecolor='black')
plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["Quantity"], bins=5, edgecolor='black')
plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

PHASE 4 : PRODUCT ANALYSIS

```python
product_revenue = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)
print(product_revenue)
```

```python
product_revenue.sort_values().plot(kind="barh",edgecolor='black')
plt.title("Revenue by Product")
plt.xlabel("Total Revenue")
plt.ylabel("Product")
plt.show()
```

```python
product_orders = df["Product"].value_counts()
print(product_orders)
```

```python
product_orders.plot(kind="bar",edgecolor='black')
plt.title("Number of Orders by Product")
plt.xlabel("Product")
plt.ylabel("Orders")
plt.xticks(rotation=45)
plt.show()
```

PHASE 5 : REFERRAL SOURCE ANALYIS

```python
referral_revenue = df.groupby("ReferralSource")["TotalPrice"].sum().sort_values(ascending=False)
print(referral_revenue)
```

```python
referral_revenue.sort_values().plot(kind="barh",edgecolor='black')
plt.title("Revenue by Referral Source")
plt.xlabel("Total Revenue")
plt.ylabel("Referral Source")
plt.show()
```

PHASE 6 : PAYMENT ANALYSIS

```python
payment_counts = df["PaymentMethod"].value_counts()
print(payment_counts)
```

```python
payment_counts.plot(kind="bar",edgecolor='black')
plt.title("Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

PHASE 7 : COUPON ANALYSIS

```python
df["UsedCoupon"] = df["CouponCode"].notna()
coupon_analysis = df.groupby("UsedCoupon")["TotalPrice"].agg(["count", "mean", "median", "sum"])
print(coupon_analysis)
```

PHASE 8: OUTLIERS DETECTION

```python
plt.boxplot(df["TotalPrice"])
plt.title("Boxplot of Total Price")
plt.ylabel("Total Price")
plt.show()
```

```python
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df["TotalPrice"] < lower_bound) |(df["TotalPrice"] > upper_bound)]
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of outliers:", len(outliers))
print(outliers[["OrderID", "Product", "Quantity", "UnitPrice", "TotalPrice"]])


```

PHASE 10 : CORRELATION ANALYSIS

```python
correlation = df[["Quantity", "UnitPrice", "ItemsInCart", "TotalPrice"]].corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
